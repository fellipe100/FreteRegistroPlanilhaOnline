```python
from core import *

def pesquisa_sap():
    print("\n\n[Step Being]--> PesquisaSAP")

    # Load memory
    in_process_folder = pieces.gvars.in_process_folder
    in_process_file = pieces.lib.get_file_from_folder(directory=in_process_folder,
                                                             endswith=".json",
                                                             not_exact_name="process_info_memory.json",
                                                             timeout_loop=1)
    memory_path = pieces.gvars.memory_full_path
    memory_dict = pieces.lib.read_json(memory_path)
    if isinstance(memory_dict, dict):
        pass
    elif memory_dict == -3:
        # Something bad happened, we shouldn't fail in here
        pieces.lib.idontknow("PesquisaSAP(): Passei do PegaTarefaPasta sem ter um arquivo de memoria na pasta 'processando', há algo de errado!")
    else:
        # Shouldnt happen, move to segregation for further analysis
        # @ TODO send an email to supporte team?
        segregation_folder = pieces.gvars.segregation_folder
        pieces.lib_processo.move_files_to_segregation(in_process_file, memory_path, segregation_folder)
        pieces.lib.iamlost("[WarnSupport] PesquisaSAP(): Memoria Corrompida")
        
    # Check which enviroment we are using
    env = pieces.gvars.enviroment
    if env == "hmg":
        sap_username, sap_password = pieces.lib.bpa_safe_get("sap_hmg")
    elif env == "hmg":
        sap_username, sap_password = pieces.lib.bpa_safe_get("sap_prd")
    else:
        pieces.lib.iamlost("pesquisa_sap(): Invalid pieces.lib.enviroment. Expected 'prd' or 'hmg' >> ".format(env))
        
    cnpj_search = memory_dict.get("cnpj", None)
    if cnpj_search is None:
        # Something bad happened, we shouldn't fail in here
        pieces.lib.idontknow("PesquisaSAP(): Passei pelo PegaTarefaPasta e na memória não consta um valor o 'CNPJ'. Há algo de errado!")
        
    numero_referencia_to_match = memory_dict.get("numero_referencia", None)
    if numero_referencia_to_match is None:
        # Something bad happened, we shouldn't fail in here
        pieces.lib.idontknow("PesquisaSAP(): Passei pelo PegaTarefaPasta e na memória não consta um valor o 'numero_referencia''. Há algo de errado!")

    c = 0
    main_timer = pieces.lib.Timer(5*60)
    while main_timer.not_expired:
        c += 1
        pieces.lib.filelog("Attempt SAP number: {}".format(c))
        # ----- Open SAP Logon, if it doesnt exists
        path_sap = pieces.gvars.path_sap
        a = open_sap_logon(path_sap)
        print("open_sap_logon --> " + str(a))

        # ----- Check for running SAP connections
        sap_session, sap_application = get_running_sap_session()
        print("get_running_sap_session --> " + str(sap_session))
        if sap_session == None:
            # ----- Start a session
            sap_session = login_sap(sap_username, sap_password, env="SAP - QA")
            print("login_sap --> " + str(sap_session))

        # ----- Go for the transaction, check if exists by accessing element "check_field_id"
        ID_FIELD_CUSTOMER_ACCOUNT = "wnd[0]/usr/ctxtDD_KUNNR-LOW"
        enter_transaction_ret = enter_transaction_sap(sap_session, "fbl5n", check_field_id=ID_FIELD_CUSTOMER_ACCOUNT)
        print("enter_transaction_ret --> " + str(enter_transaction_ret))

        # ----- Search for customer reg info
        search_for_reference_entry = search_transaction_fbl5n(sap_session, cnpj=cnpj_search, layout="ROBOBPA")
        print("search_for_reference_entry --> " + str(search_for_reference_entry))
        if search_for_reference_entry == True:
            pass
        elif search_for_reference_entry == -7:
            # Logout SAP
            enter_transaction_sap(sap_session, "ex", check_field_id=None)
            ##################################################################################
            ## PROCESS EXIT: SAP - CNPJ not found at SAP database                           ##
            ##################################################################################
            pieces.lib.filelog("SAP - CNPJ not found in SAP database. Reporting it to Dashboard")
            pieces.lib.file_delete(in_process_file)
            pieces.lib.set_obs(cnpj_search)
            pieces.lib.set_var_process("status_key", "CNPJNaoEncontradoSAP")
            pieces.lib.set_var_process("status_desc_text", "O CNPJ informado nao esta na base de dados do SAP.")
            pieces.lib.set_var_process("status_exit_code", "21")
            pieces.lib.set_var_process("obs1", cnpj_search)
            pieces.lib.idontknow("[AnaliseManual] Nenhuma conta encontrada para o CNPJ.")
        else:
            continue
            
        # ----- Capture results
        customer_info_dict = capture_results_sap(sap_session, numero_referencia_to_match, env=env)
        if isinstance(customer_info_dict, dict):
            pieces.lib.filelog("Found info about the customer! --> {}".format(customer_info_dict))
        elif customer_info_dict == -8:
            # Msg "No momento esta função encontra-se indisponível"
            ##################################################################################
            ## PROCESS EXIT: SAP - No valid accounts for the specified CNPJ                 ##
            ##################################################################################
            pieces.lib.filelog("SAP - No valid accounts for the specified CNPJ. Reporting it to Dashboard")
            pieces.lib.file_delete(in_process_file)
            pieces.lib.set_obs(cnpj_search)
            pieces.lib.set_var_process("status_key", "NenhumaContaParaOCNPJ")
            pieces.lib.set_var_process("status_desc_text", "Nenhuma conta de cliente localizada para o CNPJ informado.")
            pieces.lib.set_var_process("status_exit_code", "21")
            pieces.lib.set_var_process("obs1", cnpj_search)
            pieces.lib.idontknow("[AnaliseManual] Nenhuma conta encontrada para o CNPJ.")
        elif customer_info_dict == -9:
            # Msg "No momento esta função encontra-se indisponível"
            ##################################################################################
            ## PROCESS EXIT: SAP - No open documents were found for the specified cnpj      ##
            ##################################################################################
            pieces.lib.filelog("SAP - No open documents for the specified CNPJ. Reporting it to Dashboard")
            pieces.lib.file_delete(in_process_file)
            pieces.lib.set_obs(cnpj_search)
            pieces.lib.set_var_process("status_key", "NenhumDocumentoEmAberto")
            pieces.lib.set_var_process("status_desc_text", "Nao ha documentos em aberto para este cliente.")
            pieces.lib.set_var_process("status_exit_code", "22")
            pieces.lib.set_var_process("obs1", cnpj_search)
            pieces.lib.idontknow("[AnaliseManual] Nenhum documento encontrado na transação fbl5n.")
        else:
            continue
        
        # ------ Treat Results
        customer_info_dict = manipulate_and_check_results_sap(customer_info_dict)
        if isinstance(customer_info_dict, dict):
            pieces.lib.filelog("Dictionary successfully validated.")
        else:
            continue
            
        # add timestamp
        customer_info_dict["sap_datetime"] = str(pieces.lib.datetime.now())
        
        # attach info to memory
        memory_path = pieces.gvars.memory_full_path
        pieces.lib.write_json(memory_path, customer_info_dict, increment_keys=True)
        
        # Upload some info to Brain
        pieces.lib.set_var_process("company_code", customer_info_dict["company_code"])
        pieces.lib.set_var_process("customer_name", customer_info_dict["customer_name"])
        pieces.lib.set_var_process("payment_method", customer_info_dict["payment_method"])
        pieces.lib.set_var_process("key_ref3", customer_info_dict["key_ref3"])
        pieces.lib.set_var_process("email", customer_info_dict["email"])
        break
        
    if main_timer.expired:
        # Shouldnt happen, unless a system is offline (and even so, it should be treated elsewhere)
        # IF SAP OR ITAU ARE OFFLINE, RESET THE AUMATION THERE WITH A PROPER DASHBOARD WARNING, DO NOT LET IT TIMEOUT
        segregation_folder = pieces.gvars.segregation_folder
        pieces.lib_processo.move_files_to_segregation(in_process_file, memory_path, segregation_folder)
        pieces.lib.iamlost("[WarnSupport] PesquisaSAP(): Timeout-ed")
        
    # exit SAP transaction, login is so fast it is not worth to keep it online and risk it being unstable
    enter_transaction_sap(sap_session, "ex", check_field_id=None)
    
    pieces.lib.set_default_route("solicita_segunda_via_itau")

    print("\n\n[Step End]--> Pesquisa Sap")
    return
     
```

```python
def manipulate_and_check_results_sap(dict_sap_results):
    """
    Inputs
        dict_sap_results      # dict - Object containing all the sap results obtained from search
    Outputs
        dict                  # Success: A dictionary containing the manipulated data
        -2                    # Error: 'dict_sap_results' is invalid. Not a dict.
        -3                    # Error: 'dict_sap_results' invalid length. Expected 11 keys-value pairs.
        -4                    # Error: 'dict_sap_results['company_code']' is not amongst allowed options.
    """
    
    # debug purposes
    vars_debug = pieces.lib.get_passed_parameters(manipulate_and_check_results_sap, locals())
    
    # input validation
    if not isinstance(dict_sap_results, dict):
        pieces.lib.filelog("manipulate_and_check_results_sap({}): Error: 'dict_sap_results' is invalid. Not a dict.".format(vars_debug))
        return -2
    
    mandatory_keys = ["key_ref3", "payment_method", "email", "company_code"]
    is_dict_valid = pieces.lib_processo.dict_contains_all_keys(dict_sap_results, mandatory_keys)
    if is_dict_valid != True:
        pieces.lib.filelog("manipulate_and_check_results_sap({}): Error: 'dict_sap_results' does not contain all the mandatory keys.".format(vars_debug))
        return -3
    
    if not dict_sap_results["company_code"] in ["0500", "0600"]:
        pieces.lib.filelog("manipulate_and_check_results_sap({}): Error: 'dict_sap_results['company_code']' is not amongst allowed options.".format(vars_debug))
        return -4
    
    # Treatment #1
    # if payment_method is empty, change it to "A"
    current_payment_method = dict_sap_results["payment_method"]
    new_payment_method = "A" if current_payment_method == "" else current_payment_method
    dict_sap_results["payment_method"] = new_payment_method
    
    return dict_sap_results


def open_sap_logon(path_sap, *, timeout=60):
    """
    Inputs
        path_sap             # str Absulte path to SAP application (local or network)
    Outputs
        pywinauto obj        # Success: SAP pywinauto at login screen
        -1                   # Error: Timeout while waiting for app window
        -2                   # Error: Invalid path or app not found in the specified folder
        -3                   # Error: Invalid path (not str)
    """
    
    pieces.lib.filelog("[[ open_sap_logon ]]")

    # validate inputs
    if not isinstance(path_sap, str):
        return -3
    if not pieces.lib.os.path.exists(path_sap):
        return -2
    
    timer = pieces.lib.Timer(timeout)
    while timer.not_expired:
        
        # Verifying whether a SAP logon window already exists
        try:
            sap_app = pieces.pywinauto.application.Application(backend='uia').connect(title_re='.*SAP Logon.*')
            sap_app_top_window = sap_app.top_window()
            return sap_app_top_window
        except pieces.pywinauto.application.findwindows.ElementNotFoundError:
            pass
    
        # Otherwise, open a new application
        pieces.lib.filelog("SAP Logon not found. Starting it.")
        try:
            pieces.subprocess.Popen(path_sap, shell=True)
            sap_app = pieces.pywinauto.Desktop(backend="uia").window(title_re='.*SAP Logon.*')
            sap_app.wait(wait_for='exists', timeout=timeout)
        except pieces.pywinauto.application.AppStartError:
            pieces.lib.ultradebug("open_sap_logon(): The specified path no longer exists or app couldn't be launched. Path -> " + str(path_sap))
            return -2
        except pieces.pywinauto.timings.TimeoutError:
            # Ignore it, let the class Timer handle the timeout (return -1)
            pass
            
    return -1


def login_sap(username, password, env, *, timeout=60):
    """
    Inputs
        username        # str SAP Username
        password        # str SAP Password
        env             # str SAP Connection enviroment ["SAP - QA", "SAPPRD - 01"]
        
    Outputs
        SAP obj         # Success: SAP COM Object
        -10             # Error: Timeout ao tentar preencher o login
        -11             # Error: Timeout ao carregar a próxima tela (tela inicial SAP)
        -2              # Error: Invalid env variable. Not "str"
        -3              # Error: Invalid env variable. Should be "SAP - QA" or "SAPPRD - 01" (case-insensitive)
        -4              # Error: SAP Enviroment is offline, unreachable or unavailable
    """
    
    # Input validation
    if not isinstance(username, str):
        pieces.lib.iamlost("login_sap(): Invalid username >> " + str(username))
    if not isinstance(password, str):
        pieces.lib.iamlost("login_sap(): Invalid password >> " + str(password))
    
    try:
        env = env.upper()
    except:
        return -2
    if env not in ["SAP - QA", "SAPPRD - 01"]:
        pieces.lib.msgbox("Attention!\n\nlogin_sap() 'env' is invalid >> {}\n\nIt wont fix itself!".format(env))
        return -3
    
    timer = pieces.lib.Timer(timeout)
    while timer.not_expired:
        SapGuiAuto = pieces.win32com.client.GetObject('SAPGUI')
        if not type(SapGuiAuto) == pieces.win32com.client.CDispatch:
            return -5

        application = SapGuiAuto.GetScriptingEngine
        
        if not type(application) == pieces.win32com.client.CDispatch:
            SapGuiAuto = None
            return -6
        
        try:
            connection = application.OpenConnection("SAP - QA", True)
        except win32com.client.dynamic.pythoncom.com_error as e:
            ## SAP env is offline / unavailable
            if "-2147352567" in str(e):
                return -6
            else:
                return -7
            
        if not type(connection) == pieces.win32com.client.CDispatch:
            application = None
            SapGuiAuto = None
            return -8

        session = connection.Children(0)
        if not type(session) == pieces.win32com.client.CDispatch:
            connection = None
            application = None
            SapGuiAuto = None
            return -9

        MAIN_SAP_WINDOW = "wnd[0]"
        session.findById(MAIN_SAP_WINDOW).maximize
        
        ID_FIELD_USERNAME = "wnd[0]/usr/txtRSYST-BNAME"
        ID_FIELD_PASSWORD = "wnd[0]/usr/pwdRSYST-BCODE"
        
        # Enter credentials
        session.findById(ID_FIELD_USERNAME).text = username
        session.findById(ID_FIELD_PASSWORD).text = password
        
        # Enter
        session.findById("wnd[0]").sendVKey(0)
        
        # try to enter any text @ transaction edit field, if it works, we are online
        ID_FIELD_TRANSACTION = "wnd[0]/tbar[0]/okcd"
        text = session.findById(ID_FIELD_TRANSACTION).text = "CHECK"
        session.findById(ID_FIELD_TRANSACTION).text = ""
        if text == "CHECK":
            return session

    return -10


def enter_transaction_sap(session, transaction_code, check_field_id):
    """
    Inputs
        session               # SAPGUI COM Object
        transaction_code      # str SAP Username
        check_field_id        # str SAP Password

    Outputs
        -10             # Error: Timeout ao tentar preencher o login
        -11             # Error: Timeout ao carregar a próxima tela (tela inicial SAP)
        -2              # Error: Invalid 'transaction_code'. Not "str"
        -3              # Error: Invalid 'check_field_id'. Not "str" and not None
        -4              # Error: Invalid 'check_field_id'. Does not contain a valid ID
    """

    pieces.lib.filelog("[[ enter_transaction ]]")
    # Validate Input
    if not isinstance(transaction_code, str):
        return -2
    if check_field_id is not None:
        if not isinstance(check_field_id, str):
            return -3
        if not "wnd[0]/usr" in check_field_id:
            return -4
    
    # If the transaction does not start with \n, attach it
    # \n makes any transaction reachable from within other transaction
    # By doing it so, we dont need SAP to be on its initial screen
    transaction_code = transaction_code.strip()
    if not transaction_code.startswith("/n"):
        transaction_code = "/n" + transaction_code
    
    session.findById("wnd[0]/tbar[0]/okcd").text = transaction_code
    session.findById("wnd[0]").sendVKey(0)
    
    # check if check_field_exists (ie: could access transaction)
    if check_field_id is not None:
        session.findById(check_field_id).text = "aa"
        session.findById(check_field_id).text = ""

    return 1
    
    
def search_transaction_fbl5n(session, cnpj, layout="MYRTES", company_code_from="0500", company_code_to="0600", timeout=30):
    """
    Inputs
        session               # SAPGUI COM Object
        cnpj                  # str - CNPJ criteria, expected len == 14
        layout                # str - layout to use in the query
        company_code_from     # str - Company code criteria start range (should be 0500 [Leao])
        company_code_to       # str - Company code criteria end range (should be 0600 [Trop])
    Outputs
        -1                    # Error: Timeout while trying to do the cnpj search/execute F8
        -2                    # Error: Invalid CNPJ. Not str.
        -3                    # Error: Invalid CNPJ length (expected: 14).
        -4                    # Error: Invalid 'company_code_from'. Not str.
        -5                    # Error: Invalid 'company_code_to'. Not str.
        -6                    # Error: Unknown SAP COM manipulation error.
        -7                    # Error: 'cnpj' not found at SAP database.
    """
  
    # debug purposes
    vars_debug = pieces.lib.get_passed_parameters(search_transaction_fbl5n, locals())
    
    # Validate input
    if not isinstance(cnpj, str):
        pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Invalid CNPJ. Not str.".format(vars_debug))
        return -2
    if len(cnpj) != 14:
        pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Invalid CNPJ length (expected: 14).".format(vars_debug))
        return -3
    
    # Treat input
    if not layout.startswith("/"):
        layout = "/" + layout
    try:
        company_code_from = str(company_code_from).zfill(4)
    except:
        pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Invalid 'company_code_from'. Not str.".format(vars_debug))
        return -4
    try:
        company_code_to = str(company_code_to).zfill(4)
    except:
        pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Invalid 'company_code_to'. Not str.".format(vars_debug))
        return -5
    
    timer = pieces.lib.Timer(timeout)
    while timer.not_expired:
        try:
            # Focus Search Help Field
            ID_FIELD_SEARCH_HELP = "wnd[0]/usr/ctxtDD_INDEX-HOTKEY"
            session.findById(ID_FIELD_SEARCH_HELP).setFocus

            # F4, enter the right-button input help
            MAIN_SAP_WINDOW = "wnd[0]"
            session.findById(MAIN_SAP_WINDOW).sendVKey(4)

            # sometimes this button is skipped because its memorized by SAP session
            try:
                session.findById("wnd[1]/usr/subSUBSCR_SELONE:SAPLSDH4:0140/sub:SAPLSDH4:0140/btnG_SELONE_STATE-BUTTON_TEXT[5,2]").press
            except pieces.win32com.client.dynamic.pythoncom.com_error as e:
                if 'The control could not be found by id.' in str(e):
                    pass
                else:
                    raise e

            ID_FIELD_TAX_CODE_1 = "wnd[1]/usr/tabsG_SELONETABSTRIP/tabpTAB001/ssubSUBSCR_PRESEL:SAPLSDH4:0220/sub:SAPLSDH4:0220/txtG_SELFLD_TAB-LOW[0,24]"
            session.findById(ID_FIELD_TAX_CODE_1).text = cnpj

            # Enter x2
            session.findById(MAIN_SAP_WINDOW).sendVKey(0)
            session.findById(MAIN_SAP_WINDOW).sendVKey(0)

            # Company Code from 0500 to 0600
            ID_FIELD_COMPANY_CODE_FROM = "wnd[0]/usr/ctxtDD_BUKRS-LOW"
            ID_FIELD_COMPANY_CODE_TO = "wnd[0]/usr/ctxtDD_BUKRS-HIGH"
            session.findById(ID_FIELD_COMPANY_CODE_FROM).text = company_code_from
            session.findById(ID_FIELD_COMPANY_CODE_TO).text = company_code_to

            # fill layout
            ID_FIELD_LAYOUT = "wnd[0]/usr/ctxtPA_VARI"
            session.findById(ID_FIELD_LAYOUT).text = layout

            # Execute F8
            try:
                session.findById(MAIN_SAP_WINDOW).sendVKey(8)
            except pieces.win32com.client.dynamic.pythoncom.com_error as e:
                exception_text = str(e)
                if 'The virtual key is not enabled.' in exception_text:
                    # If this happens, CNPJ may not exist at SAP data base
                    ID_BOTTOM_STATUS_BAR = "wnd[0]/sbar"
                    status_text = session.findById(ID_BOTTOM_STATUS_BAR).text
                    msg_number = session.findById(ID_BOTTOM_STATUS_BAR).messageNumber
                    
                    STRING_ERROR_CODE_WHEN_CNPJ_DOESNT_EXIST = "802"
                    if msg_number == STRING_ERROR_CODE_WHEN_CNPJ_DOESNT_EXIST:
                        # Send a F12 to close the menu/popup
                        session.findById(MAIN_SAP_WINDOW).sendVKey(12)
                        pieces.lib.filelog("capture_results_sap({}):\nError: 'cnpj' not found at SAP database. Full Text = {}".format(vars_debug, status_text))
                        return -7
                    
            # If it reaches here, success at executing F8
            return True
        
        except pieces.win32com.client.dynamic.pythoncom.com_error as e:
            exception_text = str(e)
            if 'The virtual key is not enabled.' in exception_text:
                continue
            else:
                pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Unknown SAP COM manipulation error. >> {}".format(vars_debug, exception_text))
                return -6
            
    pieces.lib.filelog("search_transaction_fbl5n({}):\nError: Timeout while trying to do the cnpj search/execute F8.".format(vars_debug))
    return -1


def capture_results_sap(session, reference_to_match, *, env="prd"):
    """
    Inputs
        session               # SAPGUI COM Object
        reference_to_match    # str - 'numero_referencia' obtained in the task file. Expected pattern: "000185941-1"
                                      Criteria for 'reference'. Will return the results belonging to 'reference_to_match'
        env                   # "prd", "hmg" --- If 'hmg', will return the 1st row as a match for 'reference'
                                      Reason: SAP homolog does not have up-to-date world data, so we have to fake it
    Outputs
        dict                  # Success: A dictionary containing the following 12 key,values
                                      "reference",        "company_code",  "customer_code",     "customer_name"
                                      "document_number",  "document_date", "document_due_date", "amount_currency"
                                      "payment_method",   "key_ref3",       "text_comment",      "email"
        -2                    # Error: Could not map all header fields.
        -3                    # Error: 'reference_to_match' is invalid. Not str.
        -4                    # Error: 'reference_to_match' is invalid. Not on expected pattern e.g. '000173923-1'.
        -5                    # Error: Could not find some elements in SAP results table.
        -6                    # Warning: 'reference_to_match' not found amongst the results.
        -7                    # Warning: Reached maximum line read limit.
        -8                    # Error: No accounts exists for the searched 'cnpj'.
        -9                    # Error: No results were found (no open documents).
    """
   
    # debug purposes
    vars_debug = pieces.lib.get_passed_parameters(capture_results_sap, locals())
    
    # Validate inputs
    if not isinstance(reference_to_match, str):
        pieces.lib.filelog("capture_results_sap({}):\nError: 'reference_to_match' is invalid. Not str.".format(vars_debug))
        return -3
    
    EXPECTED_LENGTH_VALID_REFERENCE_WITHOUT_HYPHEN = 10
    numero_referencia_without_hyphen = reference_to_match.replace("-", "")
    numero_referencia_treated = ''.join([c for c in numero_referencia_without_hyphen if c.isdigit()])
    if reference_to_match[9] != "-" or len(numero_referencia_treated) != EXPECTED_LENGTH_VALID_REFERENCE_WITHOUT_HYPHEN:
        pieces.lib.filelog("capture_results_sap({}):\nError: 'reference_to_match' is invalid. Not on expected pattern e.g. '000173923-1'.".format(vars_debug))
        return -4
    
    # Did the search actually bring up results?
    # If we are still the initial "fbl5n" transaction, check for possible error messages
    try:
        ID_FIELD_CUSTOMER_ACCOUNT_FROM_FBL5N = "wnd[0]/usr/ctxtDD_KUNNR-LOW"
        session.findById(ID_FIELD_CUSTOMER_ACCOUNT_FROM_FBL5N).text # just try to access it
        
        # If no exception was raised, we are still in fbl5n, lets see status bar message error
        ID_BOTTOM_STATUS_BAR = "wnd[0]/sbar"
        status_text = session.findById(ID_BOTTOM_STATUS_BAR).text
        msg_number = session.findById(ID_BOTTOM_STATUS_BAR).messageNumber
        
        STRING_ERROR_CODE_WHEN_NO_ACCOUNTS_FOR_THIS_CNPJ_EXIST = "030"
        STRING_ERROR_CODE_WHEN_NO_RESULTS_ARE_FOUND = "033"
        if msg_number == STRING_ERROR_CODE_WHEN_NO_ACCOUNTS_FOR_THIS_CNPJ_EXIST:
            # Error: CNPJ not found // No accounts fullfil the selection conditions
            pieces.lib.filelog("capture_results_sap({}):\nError: No accounts exists for the searched 'cnpj'. Full Text = {}".format(vars_debug, status_text))
            return -8
        elif msg_number == STRING_ERROR_CODE_WHEN_NO_RESULTS_ARE_FOUND:
            # Error: No results were found // No open documents
            pieces.lib.filelog("capture_results_sap({}):\nError: No results were found (no open documents). Full Text = {}".format(vars_debug, status_text))
            return -9
            
    except pieces.win32com.client.dynamic.pythoncom.com_error as e:
        pass

    # Object dict to return in this func
    obj_ret = {}
    
    # Obtain the header fields text
    header_fields_to_retrieve = {
        "company_code":  {"field_id": "wnd[0]/usr/lbl[26,2]"},
        "customer_code": {"field_id": "wnd[0]/usr/lbl[26,1]"},
        "customer_name": {"field_id": "wnd[0]/usr/lbl[26,4]"}
    }
    
    for field_name, field_attributes in header_fields_to_retrieve.items():
        try:
            id_element_sap = field_attributes["field_id"]
            obj_ret[field_name] = session.findById(id_element_sap).Text.strip()
        except pieces.win32com.client.dynamic.pythoncom.com_error as e:
            pieces.lib.filelog("capture_results_sap({}):\nError: Could not map all header fields.".format(vars_debug))
            return -2
           
    # Obtain the results from Table, iterating each Row
    fields_to_retrieve = {
        "reference"        : {"field_id": "wnd[0]/usr/lbl[4,@ROW@]"},
        "document_number"  : {"field_id": "wnd[0]/usr/lbl[17,@ROW@]"},
        "document_date"    : {"field_id": "wnd[0]/usr/lbl[28,@ROW@]"},
        "document_due_date": {"field_id": "wnd[0]/usr/lbl[39,@ROW@]"},
        "amount_currency"  : {"field_id": "wnd[0]/usr/lbl[50,@ROW@]"},
        "payment_method"   : {"field_id": "wnd[0]/usr/lbl[69,@ROW@]"},
        "key_ref3"         : {"field_id": "wnd[0]/usr/lbl[72,@ROW@]"},
        "text_comment"     : {"field_id": "wnd[0]/usr/lbl[93,@ROW@]"},
        "email"            : {"field_id": "wnd[0]/usr/lbl[144,@ROW@]"}
    }
    
    # init row is 10, then inc. by 1 (rows become 10, 11, 12, 13, 14..)
    # Iterate over SAP until no more rows are found
    INIT_ROW_NUMBER_ID_MINUS_ONE = 9
    MAX_ROW_LIMIT = 30
    # I dont think we will ever exceed 30 rows, if so, warn the support or something
    row_number = 1
    found_reference = False
    while row_number < MAX_ROW_LIMIT:
        pieces.lib.filelog("Looking for results on row {}".format(row_number))
        # Is this the correct row? Does the 'reference_to_match' and 'reference' actually match?
        REFERENCE_FIELD_ID = fields_to_retrieve["reference"]["field_id"].replace("@ROW@", str(INIT_ROW_NUMBER_ID_MINUS_ONE + row_number))
        try:
            reference_sap = session.findById(REFERENCE_FIELD_ID).Text.strip()
        except pieces.win32com.client.dynamic.pythoncom.com_error as e:
            # no more lines to read
            break
            
        # If homolog, lets assume the first row is the one we wanted (homolog env won't have up-to-date real life data)
        if env == "hmg" or reference_sap == reference_to_match:
            found_reference = True
            for field_name, field_attributes in fields_to_retrieve.items():
                try:
                    id_element_sap = field_attributes["field_id"].replace("@ROW@", str(INIT_ROW_NUMBER_ID_MINUS_ONE + row_number))
                    this_field_value = session.findById(id_element_sap).Text.strip()
                    obj_ret[field_name] = this_field_value
                except pieces.win32com.client.dynamic.pythoncom.com_error as e:
                    pieces.lib.filelog("capture_results_sap({}):\nError: Could not find some elements in SAP results table >> broke at field: {}.".format(vars_debug, field_name))
                    return -5
            break
            
        row_number += 1

    if row_number == MAX_ROW_LIMIT:
        pieces.lib.filelog("capture_results_sap({}):\nWarning: Reached maximum line read limit.".format(vars_debug))
        return -7
    
    if not found_reference:
        pieces.lib.filelog("capture_results_sap({}):\nWarning: 'reference_to_match' not found amongst the results.".format(vars_debug))
        return -6
    
    return obj_ret



    
def get_running_sap_session():
    """
    Inputs
        username        # str SAP Username
        password        # str SAP Password
        env             # str SAP Connection enviroment ["SAP - QA", "SAPPRD - 01"]
        
    Outputs
        SAP obj         # Success: SAP COM Object
        None            # Success: But SAP not found
        -2              # Error: Could not locate SAPGUI in COM lib (should NOT happen, ever)
    """
    
    pieces.lib.filelog("[[ get_running_sap_session ]]")
    
    SapGuiAuto = pieces.win32com.client.GetObject('SAPGUI')
    if not type(SapGuiAuto) == pieces.win32com.client.CDispatch:
        return -2

    # Verifying whether a SAP logon window already exists
    timer = pieces.lib.Timer(10)
    RETURN_TITLE = 0
    sap_exists = False
    try:
        sap_app = pieces.pywinauto.application.Application(backend='uia').connect(class_name='SAP_FRONTEND_SESSION')

        # Get the correct window
        # Can not be hidden (title = "") and can not be "SAP Logon"
        for w in sap_app.windows():
            window_title = w.texts()[RETURN_TITLE]
            if not "SAP Logon" in window_title and window_title != "":
                sap_exists = w
                break

        if sap_exists:
            sap_app_top_window = sap_exists
            sap_app_top_window.set_focus()
    except pieces.pywinauto.application.findwindows.ElementNotFoundError:
        pass

    application = SapGuiAuto.GetScriptingEngine

    # Return existing SAP if already online
    timer = pieces.lib.Timer(5)
    session = application.ActiveSession
    while session is None and sap_exists and timer.not_expired:
        session = application.ActiveSession
            
    return session, application

def debug_func_only_close_sap(session, application):
    session.findById("wnd[0]").close()
    session = application.ActiveSession
    session.findById("wnd[1]/usr/btnSPOP-OPTION1").press
    0/0
    return
```

<p>Cada função que realiza um trabalho significativo deverá ter seu respectivo teste unitário.

Em contradição ao jeito que programavamos em .ahk, o "sentido do desenvolvimento de código" agora deverá respeitar a metodologia de TDD (test driven development) - Primeiramente desenha-se  as funções de teste unitário e <b>somente então</b> inicia-se a codificação em si, com o o objetivo de satisfazer os testes unitários.</p>

```python
from core import *

def open_sap_logon(path_sap, *, timeout=60):
    """
    Inputs
        path_sap             # str Absulte path to SAP application (local or network)
    Outputs
        pywinauto obj        # Success: SAP pywinauto at login screen
        -1                   # Error: Timeout while waiting for app window
        -2                   # Error: Invalid path or app not found in the specified folder
        -3                   # Error: Invalid path (not str)
    """
    
    pieces.lib.filelog("[[ open_sap_logon ]]")

    # validate inputs
    if not isinstance(path_sap, str):
        return -3
    if not pieces.lib.os.path.exists(path_sap):
        return -2
    
    timer = pieces.lib.Timer(timeout)
    while timer.not_expired:
        
        # Verifying whether a SAP logon window already exists
        try:
            sap_app = pieces.pywinauto.application.Application(backend='uia').connect(title_re='.*SAP Logon.*')
            sap_app_top_window = sap_app.top_window()
            return sap_app_top_window
        except pieces.pywinauto.application.findwindows.ElementNotFoundError:
            pass
    
        # Otherwise, open a new application
        pieces.lib.filelog("SAP Logon not found. Starting it.")
        try:
            pieces.subprocess.Popen(path_sap, shell=True)
            sap_app = pieces.pywinauto.Desktop(backend="uia").window(title_re='.*SAP Logon.*')
            sap_app.wait(wait_for='exists', timeout=timeout)
        except pieces.pywinauto.application.AppStartError:
            pieces.lib.ultradebug("open_sap_logon(): The specified path no longer exists or app couldn't be launched. Path -> " + str(path_sap))
            return -2
        except pieces.pywinauto.timings.TimeoutError:
            # Ignore it, let the class Timer handle the timeout (return -1)
            pass
            
    return -1

def login_sap(username, password, env, *, timeout=60):
    """
    Inputs
        username        # str SAP Username
        password        # str SAP Password
        env             # str SAP Connection enviroment ["SAP - QA", "SAPPRD - 01"]
        
    Outputs
        SAP obj         # Success: SAP COM Object
        -10             # Error: Timeout ao tentar preencher o login
        -11             # Error: Timeout ao carregar a próxima tela (tela inicial SAP)
        -2              # Error: Invalid env variable. Not "str"
        -3              # Error: Invalid env variable. Should be "SAP - QA" or "SAPPRD - 01" (case-insensitive)
        -4              # Error: SAP Enviroment is offline, unreachable or unavailable
    """
    
    # Input validation
    if not isinstance(username, str):
        pieces.lib.iamlost("login_sap(): Invalid username >> " + str(username))
    if not isinstance(password, str):
        pieces.lib.iamlost("login_sap(): Invalid password >> " + str(password))
    
    try:
        env = env.upper()
    except:
        return -2
    if env not in ["SAP - QA", "SAPPRD - 01"]:
        pieces.lib.msgbox("Attention!\n\nlogin_sap() 'env' is invalid >> {}\n\nIt wont fix itself!".format(env))
        return -3
    
    timer = pieces.lib.Timer(timeout)
    while timer.not_expired:
        SapGuiAuto = pieces.win32com.client.GetObject('SAPGUI')
        if not type(SapGuiAuto) == pieces.win32com.client.CDispatch:
            return -5

        application = SapGuiAuto.GetScriptingEngine
        
        if not type(application) == pieces.win32com.client.CDispatch:
            SapGuiAuto = None
            return -6
        
        try:
            connection = application.OpenConnection("SAP - QA", True)
        except win32com.client.dynamic.pythoncom.com_error as e:
            ## SAP env is offline / unavailable
            if "-2147352567" in str(e):
                return -6
            else:
                return -7
            
        if not type(connection) == pieces.win32com.client.CDispatch:
            application = None
            SapGuiAuto = None
            return -8

        session = connection.Children(0)
        if not type(session) == pieces.win32com.client.CDispatch:
            connection = None
            application = None
            SapGuiAuto = None
            return -9

        MAIN_SAP_WINDOW = "wnd[0]"
        session.findById(MAIN_SAP_WINDOW).maximize
        
        ID_FIELD_USERNAME = "wnd[0]/usr/txtRSYST-BNAME"
        ID_FIELD_PASSWORD = "wnd[0]/usr/pwdRSYST-BCODE"
        ID_FIELD_LANGUAGE = "wnd[0]/usr/txtRSYST-LANGU"
        
        # Enter credentials
        session.findById(ID_FIELD_USERNAME).text = username
        session.findById(ID_FIELD_PASSWORD).text = password
        session.findById(ID_FIELD_LANGUAGE).text = "PT"
        
        # Enter
        session.findById("wnd[0]").sendVKey(0)
        
        # try to enter any text @ transaction edit field, if it works, we are online
        ID_FIELD_TRANSACTION = "wnd[0]/tbar[0]/okcd"
        text = session.findById(ID_FIELD_TRANSACTION).text = "CHECK"
        session.findById(ID_FIELD_TRANSACTION).text = ""
        if text == "CHECK":
            return session

    return -10

def enter_transaction_sap(session, transaction_code, *, check_field_id=None):
    """
    Inputs
        session               # SAPGUI COM Object
        transaction_code      # SAP transaction
        check_field_id        # verify id

    Outputs
        -10             # Error: Timeout ao tentar preencher o login
        -11             # Error: Timeout ao carregar a próxima tela (tela inicial SAP)
        -2              # Error: Invalid 'transaction_code'. Not "str"
        -3              # Error: Invalid 'check_field_id'. Not "str" and not None
        -4              # Error: Invalid 'check_field_id'. Does not contain a valid ID
    """

    pieces.lib.filelog("[[ enter_transaction ]]")
    # Validate Input
    if not isinstance(transaction_code, str):
        return -2
    if check_field_id is not None:
        if not isinstance(check_field_id, str):
            return -3
        if not "wnd[0]/usr" in check_field_id:
            return -4
    
    # If the transaction does not start with \n, attach it
    # \n makes any transaction reachable from within other transaction
    # By doing it so, we dont need SAP to be on its initial screen
    transaction_code = transaction_code.strip()
    if not transaction_code.startswith("/n"):
        transaction_code = "/n" + transaction_code
    
    session.findById("wnd[0]/tbar[0]/okcd").text = transaction_code
    session.findById("wnd[0]").sendVKey(0)
    
    # check if check_field_exists (ie: could access transaction)
    if check_field_id is not None:
        session.findById(check_field_id).text = "aa"
        session.findById(check_field_id).text = ""
    pieces.lib.filelog("[[ enter_transaction = ".format(transaction_code).format("]]"))
    return 1

def define_date_plus_interval(starting_date, interval):
    datetime = pieces.lib.datetime
    timedelta = pieces.lib.timedelta
    convert_to_datetime = lambda string_date : datetime(*list(map(int, string_date.split("/"))))
    starting_date_as_datetime = convert_to_datetime("/".join((starting_date.split("/")[::-1])))
    final_date = starting_date_as_datetime + timedelta(int(interval))
    final_date = final_date.strftime("%d.%m.%Y")
    return final_date

def search_purchase_order_non_associate(session):
    pieces.lib.filelog("search_purchase_order_non_associate : INICIO()")
    datetime = pieces.lib.datetime
    # Enconter the correct date minus 30 days
    final_date = define_date_plus_interval("07/04/2020","-30")
    pieces.lib.filelog("search_purchase_order_non_associate : data encontrada".format(final_date))
    # Puts "today -30" on the first field
    #session.findById("wnd[0]/usr/ctxtS_DT_CR-LOW").text = final_date
    session.findById("wnd[0]/usr/ctxtS_DT_CR-LOW").text = "01.01.2019"
    pieces.lib.filelog("search_purchase_order_non_associate : data inicial colocada")
    # Puts "today" on the final field
    #session.findById("wnd[0]/usr/ctxtS_DT_CR-HIGH").text = datetime.now().strftime("%d.%m.%Y")
    session.findById("wnd[0]/usr/ctxtS_DT_CR-HIGH").text = "05.01.2019"
    pieces.lib.filelog("search_purchase_order_non_associate : Data final colocada")
    # select only the right field.
    session.findById("wnd[0]/usr/chkP_ERRSEF").selected = False
    session.findById("wnd[0]/usr/chkP_CANCEL").selected = False
    session.findById("wnd[0]/usr/chkP_REJEIT").selected = False
    session.findById("wnd[0]/usr/chkP_MANUAL").selected = False
    session.findById("wnd[0]/usr/chkP_GERA").selected = False
    session.findById("wnd[0]/usr/chkP_ERRO").selected = False
    session.findById("wnd[0]/usr/chkP_SIMU").selected = False
    session.findById("wnd[0]/usr/chkP_ASSO").selected = False
    pieces.lib.filelog("search_purchase_order_non_associate : CheckBox's descelecionadas")
    
    # Populating validations - TDD - Melody #
    tdd_check = {"starting_date": session.findById("wnd[0]/usr/ctxtS_DT_CR-LOW").text,
                 "final_date": session.findById("wnd[0]/usr/ctxtS_DT_CR-HIGH").text,
                 "op_PENDE": session.findById("wnd[0]/usr/chkP_PENDE").selected,
                 "op_ERRSEF": session.findById("wnd[0]/usr/chkP_ERRSEF").selected,
                 "op_CANCEL": session.findById("wnd[0]/usr/chkP_CANCEL").selected,
                 "op_REJEIT": session.findById("wnd[0]/usr/chkP_REJEIT").selected,
                 "op_MANUAL": session.findById("wnd[0]/usr/chkP_MANUAL").selected,
                 "op_GERA": session.findById("wnd[0]/usr/chkP_GERA").selected,
                 "op_ERRO": session.findById("wnd[0]/usr/chkP_ERRO").selected,
                 "op_SIMU": session.findById("wnd[0]/usr/chkP_SIMU").selected,
                 "op_ASSO": session.findById("wnd[0]/usr/chkP_ASSO").selected
                }
    # End - Populating #
    
    # pressing the button to generate the report.
    session.findById("wnd[0]/tbar[1]/btn[8]").press()
    
    pieces.lib.filelog("search_purchase_order_non_associate : Botao para gerar relatorio pressionado")
    # Returning validations to TDD  - Melody #
    return tdd_check

# Unit Test - Melody #
def test_search_purchase_order_non_associate():
    
    # Buildup Environment
        
    
    test_cases = [{"test_name": "tarefa_correta",
                  "expected_results" : {"starting_date": "01.01.2019",
                                         "final_date": "05.01.2019",
                                         "op_PENDE": True, 
                                         "op_ERRSEF": False,
                                         "op_CANCEL": False,
                                         "op_REJEIT": False,
                                         "op_MANUAL": False,
                                         "op_GERA": False,
                                         "op_ERRO": False,
                                         "op_SIMU": False,
                                         "op_ASSO": False }
                 }]
    
    results = []
    expected_results = []
        
    for test_case in test_cases:
        results.append(search_purchase_order_non_associate(session))
        expected_results.append(test_case["expected_results"])
        
    print("unity test: {}".format(search_purchase_order_non_associate.__name__))
    print("results {}".format(results))
    print("expected_results {}".format(expected_results))
    
    assert results == expected_results
    print("Teste bem sucedido")
    return
    
def find_missing_purchase_order(session):
    pieces.lib.filelog("find_missing_purchase_order : INICIO()")
    session.findById("wnd[0]").maximize
    session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").setCurrentCell(-1,"")
    # Select all purchase order
    session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").selectAll()
    # Select the option to fill the blank pruchase order's
    session.findById("wnd[0]/mbar/menu[0]/menu[1]").select()
    while True:
        try:
          session.findById("wnd[1]").sendVKey(0)
        except:
          break
    pieces.lib.filelog("find_missing_purchase_order : FINAL()")
    return

def attach_purchase_order_request(session):
    pieces.lib.filelog("attach_purchase_order_request : INICIO()")
    # Finding the correct column
    session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "EBELN_SUGERIDO"
    # Retriving the text of the cell
    ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
    current_cell_col = session.findById(ID_TABLE).CurrentCellColumn
    i_index_current_cell_row = -1
    while True:
        i_index_current_cell_row += 1
        try:
            purchase_order_number = None
            purchase_order_number = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_cell_col)
            purchase_order_number = purchase_order_number.strip()
            # Caso encontre um pedido a atribuir realiza os passos
            if(purchase_order_number is not None and purchase_order_number != ""):
                # seleciona toda a linha do pedido
                session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").selectedRows = str(i_index_current_cell_row)
                # Reformatando o variavel
                int(i_index_current_cell_row)
                # Pressiona o botão para atribuir o pedido
                session.findById("wnd[0]/tbar[1]/btn[13]").press()
                # Popula o campo com o numero do pedido
                session.findById("wnd[1]/usr/ctxtZLES_INCTEHD01-EBELN").text = purchase_order_number
                # Pressiona o botão para confirmar a atribuição
                session.findById("wnd[1]/usr/btnBOTAO_ATRIBUIR").press()
                # após atribuir o pedido seleciona o primeiro caso de frete
                session.findById("wnd[1]/usr/cntlCONTAINER_200/shellcont/shell").currentCellColumn = ""
                # ainda seleciona o primeiro caso
                session.findById("wnd[1]/usr/cntlCONTAINER_200/shellcont/shell").selectedRows = "0"
                # Pressiona o botão para associar o processo
                session.findById("wnd[1]/tbar[0]/btn[23]").press()
                # Confirma o pop-up
                session.findById("wnd[2]/usr/btnBUTTON_1").press()
                try:
                    error_status = session.findById("wnd[2]/usr/txtMESSTXT1").text
                    session.findById("wnd[2]/tbar[0]/btn[0]").press()
                    print(error_status)
                except:
                    pass
        except pieces.win32com.client.dynamic.pythoncom.com_error as e:
            break
            
            
        ## Populating validations for TDD - Melody #
        # Select all from table and populate it into array, every order and Status Values
        tdd_check = []
        # Select all purchase order
        
        # Retriving the orders column
        session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "EBELN_SUGERIDO"
        ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
        current_order_col = session.findById(ID_TABLE).CurrentCellColumn
        
        
        # Retriving the status column
        session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "STATUS"
        ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
        current_status_col = session.findById(ID_TABLE).CurrentCellColumn
        
        i_index_current_cell_row = -1
        while True:
            i_index_current_cell_row += 1
            try:
                tdd_order_number = None
                tdd_order_number = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_order_col)
                tdd_order_number = tdd_order_number.strip()

                tdd_order_status = None
                tdd_order_status = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_status_col)
                tdd_order_status = tdd_order_status.strip()

                tdd_check.append([tdd_order_number, tdd_order_status])
            except pieces.win32com.client.dynamic.pythoncom.com_error as e:
                break
        # END Populating
        
        pieces.lib.filelog("attach_purchase_order_request : FIM()")
        
    # Returning validations to TDD - Melody # 
    return tdd_check

def simulate_shipping_fee(session):
    pieces.lib.filelog("attach_purchase_order_request : INICIO()")
    # Finding the correct column
    session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "EBELN_SUGERIDO"
    # Retriving the text of the cell
    ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
    current_cell_col = session.findById(ID_TABLE).CurrentCellColumn
    i_index_current_cell_row = -1
    while True:
        i_index_current_cell_row += 1
        try:
            purchase_order_number = None
            purchase_order_number = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_cell_col)
            purchase_order_number = purchase_order_number.strip()
            # Caso encontre um pedido a atribuir realiza os passos
            if(purchase_order_number is not None and purchase_order_number != ""):
                # Verificando se existe imposto ICMS
                taxes_column = session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "VICMS"
                taxes = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, taxes_column)
                taxes = taxes.strip()
                if(taxes is None or taxes == ""):
                    print(1)
                else:
                    # Seleciona a linha para simular o frete
                    session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").selectedRows = str(i_index_current_cell_row)
                    # Reformata a int
                    int(i_index_current_cell_row)
                    # Pressiona o botao para simular a fatura
                    session.findById("wnd[0]/tbar[1]/btn[14]").press()
                    # Coloca o primeiro texto F4
                    session.findById("wnd[1]/usr/ctxtZLES_E_LOG-IVA_ALTERADO").text = "F4"
                    # Pressiona o botao de simular
                    session.findById("wnd[1]/usr/btnSIMULAR").press()
                    all_taxes_values = get_taxes_value(session)
                    TAXES_CTE_ALIQUOTA_INDEX = 2
                    TAXES_BUYERS_ALIQUOTA_INDEX = 5
                    if(all_taxes_values[TAXES_CTE_ALIQUOTA_INDEX] == all_taxes_values[TAXES_BUYERS_ALIQUOTA_INDEX]):
                        final_price = calc_taxes_prices(session,all_taxes_values)
                        final_price_text = str(final_price)
                        session.findById("wnd[1]/usr/txtZLES_E_LOG-DESC_FRETE").text = final_price_text
                    else:
                        # Coloca o primeiro texto F3
                        session.findById("wnd[1]/usr/ctxtZLES_E_LOG-IVA_ALTERADO").text = "F3"
                        # Pressiona o botao de simular
                        session.findById("wnd[1]/usr/btnSIMULAR").press()
                        all_taxes_values = get_taxes_value(session)
                        TAXES_CTE_ALIQUOTA_INDEX = 2
                        TAXES_BUYERS_ALIQUOTA_INDEX = 5
                        if(all_taxes_values[TAXES_CTE_ALIQUOTA_INDEX] == all_taxes_values[TAXES_BUYERS_ALIQUOTA_INDEX]):
                            final_price = calc_taxes_prices(session,all_taxes_values)
                            final_price_text = str(final_price)
                            session.findById("wnd[1]/usr/txtZLES_E_LOG-DESC_FRETE").text = final_price_text
                        else:
                            set_var_process("status","Valores de imposto nao estao batendo")
                # pressiona o botao para acessar a simulacao da fatura.
        except pieces.win32com.client.dynamic.pythoncom.com_error as e:
            break   
            
        ## Populating validations for TDD - Melody #
        # Select all from table and populate it into array, every order and Status Values
        tdd_check = []
        # Select all purchase order
        
        # Retriving the orders column
        session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "EBELN_SUGERIDO"
        ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
        current_order_col = session.findById(ID_TABLE).CurrentCellColumn
        
        
        # Retriving the status column
        session.findById("wnd[0]/usr/cntlCONTAINER_100/shellcont/shell").currentCellColumn = "STATUS"
        ID_TABLE = "wnd[0]/usr/cntlCONTAINER_100/shellcont/shell"
        current_status_col = session.findById(ID_TABLE).CurrentCellColumn
        
        i_index_current_cell_row = -1
        while True:
            i_index_current_cell_row += 1
            try:
                tdd_order_number = None
                tdd_order_number = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_order_col)
                tdd_order_number = tdd_order_number.strip()

                tdd_order_status = None
                tdd_order_status = session.findById(ID_TABLE).GetCellValue(i_index_current_cell_row, current_status_col)
                tdd_order_status = tdd_order_status.strip()

                tdd_check.append([tdd_order_number, tdd_order_status])
            except pieces.win32com.client.dynamic.pythoncom.com_error as e:
                break
        # END Populating
    
    # Returning Validations for TDD #
    return tdd_check


def get_taxes_value(session):
    # aba CTe, campo valor
    cte_value = session.findById("wnd[1]/usr/txtZLES_E_LOG-VICMS").text
    #pieces.lib.msgbox("@{}@".format(cte_value))
    # aba CTe, campo Base
    cte_base_price = session.findById("wnd[1]/usr/txtZLES_E_LOG-VBC").text
   # pieces.lib.msgbox("@{}@".format(cte_base_price))
    # aba CTe, campo Aliquota
    cte_aliquota = session.findById("wnd[1]/usr/txtZLES_E_LOG-PICMS").text
   # pieces.lib.msgbox("@{}@".format(cte_aliquota))
    # aba Pedido, campo Valor
    buyers_value = session.findById("wnd[1]/usr/txtZLES_E_LOG-VICMS_PEDIDO").text
  #  pieces.lib.msgbox("@{}@".format(buyers_value))
    # aba Pedido, Base
    buyers_base_price = session.findById("wnd[1]/usr/txtZLES_E_LOG-VBC_PEDIDO").text
  #  pieces.lib.msgbox("@{}@".format(buyers_base_price))
    # aba Pedido, Aliquota
    buyers_aliquota = session.findById("wnd[1]/usr/txtZLES_E_LOG-PICMS_PEDIDO").text
  #  pieces.lib.msgbox("@{}@".format(buyers_aliquota))
    # declarando o array como um tuple
    all_taxes = (cte_value,cte_base_price,cte_aliquota,buyers_value,buyers_base_price,buyers_aliquota)
    return all_taxes

def calc_taxes_prices(session,taxes_values):
    TAXES_CTE_VALUE_INDEX = 0
    TAXES_CTE_BASE_INDEX = 1
    TAXES_BUYERS_VALUE_INDEX = 3
    TAXES_BUYERS_BASE_INDEX = 4
    
    cte_value = float(taxes_values[TAXES_CTE_VALUE_INDEX].replace(",","."))
    pieces.lib.msgbox(cte_value)
    cte_base = float(taxes_values[TAXES_CTE_BASE_INDEX].replace(",","."))
    pieces.lib.msgbox(cte_base)
    buyers_value = float(taxes_values[TAXES_BUYERS_VALUE_INDEX].replace(",","."))
    pieces.lib.msgbox(buyers_value)
    buyers_base = float(taxes_values[TAXES_BUYERS_BASE_INDEX].replace(",","."))
    pieces.lib.msgbox(buyers_base)
    
    result_cte = cte_base - cte_value
    pieces.lib.msgbox(result_cte)
    
    result_buyers = buyers_base - buyers_value
    pieces.lib.msgbox(result_buyers)
    
    final_result = result_cte - result_buyers
    final_result = round(final_result,2)
    pieces.lib.msgbox(final_result)
    

    if(final_result < 0):
        final_result = final_result * -1
        final_result = round(final_result,2)
        final_result = str("{:.2f}".format(final_result))
        final_result = final_result.replace(".",",")
        pieces.lib.msgbox(final_result)
    else:
        final_result = str("{:.2f}".format(final_result))
        final_result = final_result.replace(".",",")
    return final_result

# UNIT Test - MELODY - In Progress #
def test_attach_purchase_order_request():   
    test_cases = [{"test_name": "tarefa_correta",
                   "expected_result": {"not_valid": 0}
                  }] 
    
    results = []
    expected_results = []
    nValid = 0
    
    for test_case in test_cases:
        validator = attach_purchase_order_request(session)
        
        # Checks if orders were sucecifully associated #
        for valid in validator:  
            # IF has order and is not associated
            if (valid[1] is not None and valid[0] != "") and valid[1] != "ASSOCIADO": ## Correct associated value
#                 nValid += 1
                teste = 0 # Apenas para teste, retirar quando corrigir o valor de ASSOCIADO
            # IF doesn't have order and is not Pending
            else:
                if (valid[1] is not None and valid[0] == "") and valid[1] != "@0A@": ## @0A@ == Pending
                    nValid += 1
 
        results.append({"not_valid" : nValid})
        expected_results.append(test_case["expected_result"])  

    print("unity test: {}".format(attach_purchase_order_request.__name__))
    print("results {}".format(results))
    print("expected_results {}".format(expected_results))

    assert results == expected_results

    print("Teste bem sucedido")
    
    return

# Unit Test - Melody - In Progress #
def test_simulate_shipping_fee():
    test_cases = [{"test_name": "tarefa_correta",
                   "expected_result": {"not_valid": 0}
                  }] 
    
    results = []
    expected_results = []
    nValid = 0
    
    for test_case in test_cases:
        validator = simulate_shipping_fee(session)      
        # Checks if orders were sucecifully associated and simulated #
        for valid in validator:          
            # IF has order and is not associated and simulated
            if (valid[1] is not None and valid[0] != "") and valid[1] != "@08@": ## @08@ == Associated and simulated
                nValid += 1
            # IF doesn't have order and is not Pending
            else:
                if (valid[1] is not None and valid[0] == "") and valid[1] != "@0A@": ## @0A@ == Pending
                    nValid += 1
 
        results.append({"not_valid" : nValid})
        expected_results.append(test_case["expected_result"])  

    print("unity test: {}".format(simulate_shipping_fee.__name__))
    print("results {}".format(results))
    print("expected_results {}".format(expected_results))

    assert results == expected_results

    print("Teste bem sucedido")
    
    return


# Unit Test - Melody #
def test_calc_taxes_prices():
    
    test_cases = [{"test_name": "aliquota_igual",
                   "taxes_values": ["10,00", # cte_value,
                                    "20,55", # cte_base_price,
                                    "30,00", # cte_aliquota,
                                    "10,00", # buyers_value,
                                    "20,00", # buyers_base_price,
                                    "30,00", # buyers_aliquota
                                   ],
                   "expected_result": "0,55"},
                  {"test_name": "aliquota_diferente",
                   "taxes_values": ["15,00", # cte_value,
                                    "50,00", # cte_base_price,
                                    "30,00", # cte_aliquota,
                                    "23,00", # buyers_value,
                                    "45,00", # buyers_base_price,
                                    "65,00", # buyers_aliquota
                                   ],
                   "expected_result": "13,00"}]
    results = []
    expected_results = []
    
    for test_case in test_cases:
        results.append(calc_taxes_prices(session, test_case["taxes_values"]))
        expected_results.append(test_case["expected_result"])  

    print("unity test: {}".format(calc_taxes_prices.__name__))
    print("results {}".format(results))
    print("expected_results {}".format(expected_results))

    assert results == expected_results

    print("Teste bem sucedido")
    return



# permitir os testes de serem rodados somente via Jupyter
if isDebug(__name__):
#     open_sap_logon("C:\Program Files (x86)\SAP\FrontEnd\SAPgui\saplogon.exe")
#     session = login_sap("usr_robos","Bpa@1234*","SAP - QA")
#     enter_transaction_sap(session,"/nZLES010T")
#     test_search_purchase_order_non_associate()
#     find_missing_purchase_order(session)
    #session, whatever = get_running_sap_session()
#     test_attach_purchase_order_request()
    
#     test_calc_taxes_prices()
    test_simulate_shipping_fee()
    pass
```

```python
###### ---------->>>> Don't forget to install gspread and oauth2client --> pip install gspread oauth2client
import gspread 
from oauth2client.service_account import ServiceAccountCredentials

#Opens a worksheet via API at websheets Google
def openWebSheet(authFile, worksheet, activeSheet=''):

    # authFile                 =    name of authentication file provided by Google API Service
    # worksheet                =    name of the websheet
    # activeSheet (optional)   =    title of the sheet to be activated and returned. If empty, returns default sheet  
    
    scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
    creds = ServiceAccountCredentials.from_json_keyfile_name(authFile, scope)
    client = gspread.authorize(creds)
    
    # Opens the conection
    ws = client.open(worksheet)
    
    # Checks if the sheet title was passed to the function
    if activeSheet != '':
        return ws.worksheet(activeSheet)      
    
    #Returns the worksheet on the default sheet
    return ws.sheet1 


## CHEATSHEET ##
#
## -- READING -- ##
#
# -- Returns every row inside an array (inside and array) - EX: [{'pedido': 1234, 'frete': 11, 'status': 'ok'}]
# sheet.get_all_values()
#
# -- Returns every row inside an object (inside and array) - EX: [['pedido', 'frete', 'status'], ['1234', '11', 'ok']]
# sheet.get_all_records()
#
# -- Returns values of a specific row
# ROW_INDEX = 2
# sheet.row_values(ROW_INDEX)
#
# -- Returns values of a specific column
# COLUMN_INDEX = 5
# sheet.col_values(COLUMN_INDEX)
#
# -- Returns a particular cell value 
# ROW_INDEX = 1
# COLUMN_INDEX = 2
# sheet.cell(ROW_INDEX, COLUMN_INDEX).value
#
## -- WRITING -- ##
#
# -- Inserting a new row (It creates a new row, it does not update an existing row)
# ROW_VALUES = ["12345", "10,00", "Aguardando"] 
# ROW_INDEX = 2 --------- > New row will be created at this index, pushing any existing rows down 
# sheet.insert_row(ROW_VALUES, ROW_INDEX)
#
# -- Updating the value of a specific cell
# ROW_INDEX = 1
# COLUMN_INDEX = 2
# CELL_VALUE = 'observacao'
# sheet.update_cell(ROW_INDEX, COLUMN_INDEX, CELL_VALUE)


# Testing the function
sheet = openWebSheet('mytest-273716-aaddc43030cc.json', 'meuTeste') #file on folder src/ipynb
print(sheet.title)
print(sheet.get_all_records())


```
